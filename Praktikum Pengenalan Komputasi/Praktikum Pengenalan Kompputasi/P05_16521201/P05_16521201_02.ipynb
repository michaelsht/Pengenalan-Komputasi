```python
# Nomor 1
# Banyaknya pengguna tidak aktif yang berusia di atas 50 tahun
import pandas as pd

# Pandas akan membaca file P05-01.csv yang berada di folder yang sama dengan file ipynb ini
df = pd.read_csv("P05-01.csv")
df1 = df.loc[df["Attrition_Flag"] == "Attrited Customer"]
A = df1.loc[df1["Customer_Age"] > 50]

# Menampilkan banyaknya pengguna tidak aktif yang berusia di atas 50 tahun
print("Banyaknya pengguna tidak aktif yang berusia di atas 50 tahun: ", len(A))
```

```python
# Nomor 2
# ID pengguna dengan nominal limit kartu kredit terbesar
import pandas as pd

# Pandas akan membaca file P05-01.csv yang berada di folder yang sama dengan file ipynb ini
df2 = pd.read_csv("P05-01.csv")
kredit_max = df2["Credit_Limit"].max()

# Menampilkan ID pengguna dengan nominal limit kartu kredit terbesar
df2.loc[df2["Credit_Limit"] == kredit_max]
```

```python
# Nomor 3
# Banyaknya pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40
import pandas as pd

# Pandas akan membaca file P05-01.csv yang berada di folder yang sama dengan file ipynb ini
df3 = pd.read_csv("P05-01.csv")
B= df3.loc[(df3["Card_Category"] == "Blue") & (df3["Months_on_book"] > 40)]
# Menampilkan banyaknya pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40
print("banyaknya pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40: ", len(B))
```

```python
# Nomor 5
# 5 ID pengguna aktif dengan banyak bulan kartu kredit tidak aktif dalam 12 bulan terakhir terbanyak.
import pandas as pd

# Pandas akan membaca file P05-01.csv yang berada di folder yang sama dengan file ipynb ini
df = pd.read_csv("P05-01.csv")

# Menampilkan 5 ID pengguna aktif dengan banyak bulan kartu kredit tidak aktif dalam 12 bulan terakhir terbanyak.
df5 = df.loc[df["Attrition_Flag"] == "Existing Customer"]
df5.sort_values(["Months_Inactive_12_mon"], ascending=[False])[:5]
```
