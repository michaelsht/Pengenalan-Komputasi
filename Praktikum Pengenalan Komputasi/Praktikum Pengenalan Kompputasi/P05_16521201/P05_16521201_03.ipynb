```python
# Nomor 1
# Histogram distribusi banyaknya review rumah yang disewakan di wilayah west Region dengan bins 20
import pandas as pd

# Pandas akan membaca file P05-03.csv yang berada di folder yang sama dengan file ipynb ini
df1 = pd.read_csv("P05-03.csv")

# Menampilkan Histogram distribusi banyaknya review rumah yang disewakan di wilayah west Region dengan bins 20
df1.loc[df1["neighbourhood_group"]=="West Region"][["number_of_reviews"]].plot(kind="hist",bins = [0,20,40,60,80,100])
```

```python
# Nomor 2 
# Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam diatas 300

# Pandas akan membaca file P05-03.csv yang berada di folder yang sama dengan file ipynb ini
df2 = pd.read_csv("P05-03.csv")

# Menampilkan Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam diatas 300
df2.loc[df2["price"]>300]["neighbourhood_group"].value_counts().plot(kind="pie")
```

```python
# Nomor 3
# Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang

# Pandas akan membaca file P05-03.csv yang berada di folder yang sama dengan file ipynb ini
df3 = pd.read_csv("P05-03.csv")

# Menampilkan hasil
df3.loc[df3["neighbourhood"]=="Kallang"].plot(kind="scatter", x = "longitude", y = "latitude")
```

```python
# Nomor 4
# Pandas akan membaca file P05-03.csv yang berada di folder yang sama dengan file ipynb ini
df4 = pd.read_csv("P05-03.csv")

# Menampilkan hasil
df4.loc[df4["availability_365"]>100].plot(kind="scatter", x = "longitude", sizes = df4["number_of_reviews"], y = "latitude")
```

```python
# Nomor 5
# Pandas akan membaca file P05-03.csv yang berada di folder yang sama dengan file ipynb ini
df5 = pd.read_csv("P05-03.csv")

# Menampilkan hasil
df5.groupby(["neighbourhood", "room_type"]).size().unstack().plot(kind="bar", stacked= True)
```
