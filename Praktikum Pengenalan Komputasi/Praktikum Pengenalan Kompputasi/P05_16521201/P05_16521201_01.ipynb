```python
# Nomor 2
# Pajak bangunan per tahun terbesar
import pandas as pd

# Pandas akan membaca file P05-02.csv yang berada di folder yang sama dengan file ipynb ini
df2 = pd.read_csv("P05-02.csv")

# Menampilkan pajak bangunan per tahun terbesar
pajak_max = df2["property_tax"].max()
df2.loc[df2["property_tax"] == pajak_max]
```

```python
# Nomor 3
# 5 rumah dengan harga sewa per tahun terbesar
import pandas as pd

# Pandas akan membaca file P05-02.csv yang berada di folder yang sama dengan file ipynb ini
df3 = pd.read_csv("P05-02.csv")

# Menampilkan 5 rumah dengan harga sewa per tahun terbesar
df3.sort_values(["rent_amount"], ascending = [False])[:5]
```

```python
# Nomor 5
# Nilai korelasi antara jumlah tempat parkir dengan pajak tanah per tahun
import pandas as pd

# Pandas akan membaca file P05-02.csv yang berada di folder yang sama dengan file ipynb ini
df5 = pd.read_csv("P05-02.csv")

# Menampilkan nilai korelasi antara jumlah tempat parkir dengan pajak tanah per tahun
print("Tingkat Korelasi antara jumlah tempat parkir dengan pajak tanah pertahun")
df5["parking_spaces"].corr(df5["homeowner_tax"])
```
